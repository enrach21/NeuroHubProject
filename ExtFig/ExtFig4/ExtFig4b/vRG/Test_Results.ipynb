# MG

```r
pos <- read.table('MG_Pos_chr2_predict_output.txt')
dim(pos)
head(pos)
```

```r
neg <- read.table('MG_Neg_chr2_predict_output.txt')
dim(neg)
head(neg)
```

```r
pr <- pr.curve(pos$V2, neg$V2, curve=T)
plot(pr)
roc <- roc.curve(pos$V2, neg$V2, curve=T)
plot(roc)
```
