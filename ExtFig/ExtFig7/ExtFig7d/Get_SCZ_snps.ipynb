# Make the shuffled regions

```r
df <- read.csv('Scz.Filt.allATAC.SNPs.5.1.24.csv')
head(df)
dim(df)
```

```r
df <- read.csv('Scz.Filt.allATAC.SNPs.200bp.5.1.24.csv', row.names=1)
head(df)
dim(df)
```

```r
# Get ref allele
ref <- df$ref
ref
length(ref)

# Get alt allele
alt <- df$alts
alt
length(alt)
```

```r
shuf <- read.fasta('Scz.Shuffled.SNPs.5.1.24.fa')
length(shuf)
```

```r
shuf <- read.fasta('Scz.Shuffled.SNPs.200bp.5.1.24.fa')
length(shuf)
```

```r
fasta_alt <- shuf
fasta_ref <- shuf
num <- 3
for (x in 1:length(ref)){
    for (y in 1:num) {
        z <- ((x-1)*num)+y

        fasta_alt[[z]][500:(500 + nchar(alt[x]) - 1)] <- str_split(alt[x], pattern = "")[[1]]
    
        fasta_ref[[z]][500:(500 + nchar(ref[x]) - 1)] <- str_split(ref[x], pattern = "")[[1]]
    
        
    }
}

write.fasta(fasta_alt,make.unique(names(fasta_alt)), 'Scz_Alt_Shuffled.SNPs.5.1.24.fa')
write.fasta(fasta_ref,make.unique(names(fasta_ref)), 'Scz_Ref_Shuffled.SNPs.5.1.24.fa')
```

```r
fasta_alt <- shuf
fasta_ref <- shuf
num <- 3
for (x in 1:length(ref)){
    for (y in 1:num) {
        z <- ((x-1)*num)+y

        fasta_alt[[z]][100:(100 + nchar(alt[x]) - 1)] <- str_split(alt[x], pattern = "")[[1]]
    
        fasta_ref[[z]][100:(100 + nchar(ref[x]) - 1)] <- str_split(ref[x], pattern = "")[[1]]
    
        
    }
}

write.fasta(fasta_alt,make.unique(names(fasta_alt)), 'Scz_Alt_Shuffled.SNPs.200bp.5.1.24.fa')
write.fasta(fasta_ref,make.unique(names(fasta_ref)), 'Scz_Ref_Shuffled.SNPs.200bp.5.1.24.fa')
```
